```python
#CELL 1
import math as mp
import numpy as np
import vpython as vp
from vpython import color, rate,arrow,label
import matplotlib.pylab as py
import animations as ani
```

```python
#CELL 2
#This animates a rotating ring or returns graphs based on the parameter inputs.
gravity = input("What is your desired gravity in units of g?")
radius = input("What is your desired radius in meters?")
omega = input("What is your desired angular velocity in rad/s?")
if str(omega) + str(radius) + str (gravity) == "":
    print("Invalid Input Values, Please Try Again")
elif str(gravity) + str(radius) == "":
    gravity_values = []
    radius_values = []
    for i in np.linspace(0.1,20,50):
        gravity_values.append(i)
    for ii in gravity_values:
        radius_values.append(ani.radius_calc(ii,omega))
    py.plot(gravity_values,radius_values)
    py.xlabel('Gravity (g)')
    py.ylabel('Radius (m)')
    py.title('Radius (m) vs Gravity (g) for given omega '+omega+'rad/s')
    py.show()
elif str(gravity) + str(omega) == "":
    gravity_values_1 = []
    omega_values = []
    for j in np.linspace(0.1,50,50):
        omega_values.append(j)
    for jj in omega_values:
        gravity_values_1.append(ani.gravity_calc(radius,jj))
    py.plot(omega_values,gravity_values_1)
    py.xlabel('Omega (rad/s)')
    py.ylabel('Gravity (g)')
    py.title('Gravity (g) vs Omega (rad/s) for radius '+radius+'m')
    py.show()
elif str(omega) + str(radius) == "":
    omega_values_1 = []
    radius_values_1 = []
    for k in np.linspace(0.1,1000,1000):
        radius_values_1.append(k)
    for kk in radius_values_1:
        omega_values_1.append(ani.omega_calc(gravity, kk))
    py.plot(radius_values_1,omega_values_1)
    py.xlabel('Radius (m)')
    py.ylabel('Omega (rad/s)')
    py.title('Omega (rad/s) vs Radius (m) for Given Gravity '+ gravity+'g')
    py.show()
elif gravity == "":
    new_gravity = (ani.gravity_calc(radius, omega))
    print("Gravity is ",new_gravity,"g")
    print("The Period of Rotation is ", (2*mp.pi)/float(omega),"s")
    ani.ring_animation(float(radius),float(omega))
elif radius == "":
    new_radius = (ani.radius_calc(gravity, omega))
    print("Radius is ",new_radius, "m") 
    print("The Period of Rotation is ", (2*mp.pi)/float(omega),"s")
    ani.ring_animation(new_radius,float(omega))
elif omega == "":
    new_omega = (ani.omega_calc(gravity,radius))
    print("Angular Velocity is ",new_omega, "rad/s")
    print("The Period of Rotation is ", (2*mp.pi)/float(new_omega),"s")
    ani.ring_animation(float(radius),new_omega)
else:
    print("Invalid Input Values, Please Try Again")
```

```python
#CELL 3
#This animates the projectile motion of an object from the rotating reference frame.
radius_c = float(input("What is your given radius in meters? "))
omega_c = float(input("What is your angular velocity (omega) ? "))
added_throw = float(input("What is the added velocity(m/s) of the object? "))
angle_launch = float(input("What is the angle of your launch in degrees? "))
spin = float(input("What is the direction of your throw? (1 for spinward and -1 for anti-spinward) "))

c = vp.canvas()
ani.plot_throw(radius_c,omega_c,added_throw,angle_launch,spin)
py.show()
print("Animation simulates 1/10 of a second per second.")
ani.throw_animation(radius_c,omega_c,added_throw,angle_launch,spin)
```

```python
#CELL 4
#This plots the trajectories of a range of velocities added to the throw of the object from velocities between 0m/s and 10 m/s.
for i in np.linspace(0,10,5):
    ani.plot_throw(radius_c,omega_c,i,angle_launch,spin)
py.show()
```

```python
#CELL 5
#This plots the trajectories of a range of launch angles between 0 and 90 degrees.
for j in np.linspace(0,90,5):
    ani.plot_throw(radius_c,omega_c,added_throw,j,spin)
py.show()
```
